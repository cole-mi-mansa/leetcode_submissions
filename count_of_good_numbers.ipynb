```python
def n_digit_numbers(n):
    n_arr = []
    for i in range((10**(n-1)), 10**n):
        n_arr.append(i)
    return n_arr

def palindrome_num(n):
    palindrome = []
    for i in n_digit_numbers(n):
        arr = []
        for j in str(i):
            arr.append(int(j))
        reversed_arr = arr[::-1]
        if arr == reversed_arr:
            palindrome.append(i)
    return palindrome

def shift_palindrome(n, x):
    shift_pal = []
    count = 0
    for i in k_palindromic(n, x):
        pal = []
        for j in str(i):
            pal.append(int(j))
        for k in n_digit_numbers(n):
            num = []
            for l in str(k):
                num.append(int(l))
            exist = 0
            for o in num:
                for m in range(n):
                    if pal[m] == o:
                        exist += 1
            if exist == n+2:
                print(k)
                count += 1
                shift_pal.append(k)
    for y in k_palindromic(n, x):
        if y in shift_pal:
            continue
        else:
            print(y)
            shift_pal.append(y)
            count += 1
    return f"There are {count} good numbers with {n} digits and k as {x}"
        

def k_palindromic(n, k):
    k_pal = []
    for i in palindrome_num(n):
        if i % k == 0:
            k_pal.append(i)
    return k_pal

```
